```python
def add(num1, num2):
  return num1 + num2
def subtract(num1, num2):
  return num1 - num2
def multiply(num1, num2):
  return num1 * num2
def divide(num1, num2):
  if num2 == 0:
    return "Error! Division by Zero."
  return num1 / num2
print(" Welcome To your core Math Engine ")
while True:
  print("\n--- MENU OPTIONS---")
  print("1. Add (+)")
  print("2. subtract (-)")
  print("3. Multiply (*)")
  print("4. Divide (/)")
  print("5. Exit Calculator")
  choice = input("Select an option (1-5): ").strip()
  if choice == "5":
    print("closing math engine. Goodbye! ")
    break
  if choice in ["1", "2", "3", "4"]:
    a = float(input("Enter first number: "))
    b = float(input("Enter second number: "))
    if choice == "1":
      result = add(a, b)
      print(f" Result: {a} + {b} = {result}")
    elif choice == "2":
      result = subtract(a, b)
      print(f" Result: {a} - {b} = {result}")
    elif choice == "3":
      result = multiply(a, b)
      print(f" Result: {a} * {b} = {result}")
    elif choice == "4":
      result = divide(a, b)
      print(f" Result: {result}")
    else:
      print(" Invalid selection! Please choose a number from 1 to 5.")
```
